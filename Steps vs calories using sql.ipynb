```python
#--- Import Pandas ---
import pandas as pd
#--- Read in dataset ----
df = pd.read_csv('dailyActivity.csv')

# ---WRITE YOUR CODE FOR TASK 1 ---
#--- Inspect data --
df.info()
df.head()
df.describe()
```

```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Verify removal of duplicates
num_duplicates_after = df_cleaned.duplicated().sum()
print(f"Number of duplicate rows after cleaning: {num_duplicates_after}")

# Optional: Save the cleaned data to a new CSV file
df_cleaned.to_csv('dailyActivity_cleaned.csv', index=False)
```

```python
null_counts = df.isnull().sum()

# Display the null counts
print("Count of null values in each column:")
print(null_counts)
```

```python
import numpy as np
from scipy import stats
import pandas as pd
#--- WRITE YOUR CODE FOR TASK 4 ---
#df = ...

#--- Inspect data ---
z_scores_total_steps = np.abs(stats.zscore(df['TotalSteps']))
z_scores_calories = np.abs(stats.zscore(df['Calories']))

# Define a threshold for identifying outliers
threshold = 3

# Identify outliers
outliers_total_steps = np.where(z_scores_total_steps > threshold)
outliers_calories = np.where(z_scores_calories > threshold)

# Display the indices of outliers
print("Indices of outliers in 'TotalSteps':", outliers_total_steps[0])
print("Indices of outliers in 'Calories':", outliers_calories[0])

# Inspect data - Display rows with outliers in 'TotalSteps' and 'Calories'
print("\nOutliers in 'TotalSteps':")
print(df.iloc[outliers_total_steps])

print("\nOutliers in 'Calories':")
print(df.iloc[outliers_calories])

# Optional: Remove outliers to clean the dataset
df_cleaned = df[(z_scores_total_steps <= threshold) & (z_scores_calories <= threshold)]

# Inspect the cleaned data
print("\nCleaned Data:")
print(df_cleaned.head())
```

```python
#--- WRITE YOUR CODE FOR TASK 6 ---
#df = ...

# Extract the day of the week from 'ActivityDate'
df['Day'] = df['ActivityDate'].dt.day_name()

# Inspect the data
print("Data with 'Day' column added:")
print(df.head())
```

```python
#--- WRITE YOUR CODE FOR TASK 7 ---
#df = ...

# Calculate 'TotalMinutes' by summing up various activity minutes
df['TotalMinutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']

# Inspect the data
print("Data with 'TotalMinutes' column added:")
print(df.head())
```

# A Covenant with 10,000 Strides.
In our data covenant, this SQL scripture unveils as a sentinel, identifying the stalwarts committed to a daily pilgrimage of 10,000 steps within 'smart_watch_data.' By grouping 'Id'entities and comparing the count of distinct 'Days' to the days surpassing the 10,000-step threshold, the code illuminates those who consistently meet this wellness milestone. It's not just a query; it's a proclamation of dedication to an active lifestyle. This code becomes a herald, singling out the individuals whose footsteps echo the rhythm of sustained commitment, painting a portrait of steadfast wellness within our diverse community.

```python
%%sql
SELECT Id,
       COUNT(DISTINCT CASE WHEN TotalSteps >= 10000 THEN Day ELSE NULL END) AS days_over_10000_steps,
       COUNT(DISTINCT Day) AS total_active_days,
       ROUND((COUNT(DISTINCT CASE WHEN TotalSteps >= 10000 THEN Day ELSE NULL END) / COUNT(DISTINCT Day) * 100), 2) AS percentage_over_10000
FROM smart_watch_data
GROUP BY Id
HAVING total_active_days > 0
ORDER BY percentage_over_10000 DESC;
```

#  Averaging Life's Active Minutes.
In our temporal expedition, this SQL script emerges as a timekeeper, computing the average daily minutes spent in various activity levels within 'smart_watch_data.' By grouping 'Id'entities, the code creates a symphony of average minutes, portraying the nuanced cadence of each individual's daily routine. It's not merely a calculation; it's a narrative of how minutes unfold into a mosaic of activity. This code becomes a storyteller, weaving a tale of diverse lifestyles and the shared heartbeat of active minutes, fostering a deeper understanding of the collective rhythms that shape our wellness narrative.

```python
%%sql
SELECT Id,
       AVG(VeryActiveMinutes) AS avg_very_active_minutes,
       AVG(FairlyActiveMinutes) AS avg_fairly_active_minutes,
       AVG(LightlyActiveMinutes) AS avg_lightly_active_minutes,
       AVG(SedentaryMinutes) AS avg_sedentary_minutes
FROM smart_watch_data
GROUP BY Id
ORDER BY Id;
```

# Crafting Steps into Minutes.
In our data chronicle, this SQL script unfolds as an alchemist, transmuting steps into a temporal currency within 'smart_watch_data.' By grouping 'Id'entities and 'Days,' the code unveils the day-by-day maestros who maximize their step-to-minute efficiency. It's more than just a ratio; it's a story of how individuals orchestrate their movements with precision. This code becomes a storyteller, narrating the tales of those who tread the path of efficiency, painting a canvas of wellness where steps and minutes harmonize to create a symphony of purposeful activity in the diverse tapestry of our daily lives.

```python
%%sql
SELECT Id, Day,
       SUM(TotalSteps) AS TotalSteps,  -- Sum of steps for the day per ID
       SUM(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes) AS TotalActiveMinutes,
       CASE WHEN SUM(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes) > 0 
            THEN ROUND(SUM(TotalSteps) / SUM(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes), 2) 
            ELSE 0 
       END AS StepsPerMinuteEfficiency
FROM smart_watch_data
GROUP BY Id, Day
ORDER BY Day, StepsPerMinuteEfficiency DESC;
```

# Mapping Daily Distance Variation.
In our data cartography, this SQL quest unfolds as a navigator, mapping the undulating terrains of daily distance variation within 'smart_watch_data.' By encapsulating the maximum and minimum distances for each 'Id' on each 'Day,' the code illuminates the explorers whose journeys span the widest distances. It's not just a calculation; it's a tale of the varied landscapes traversed daily. This code becomes a storyteller, sketching the contours of daily adventures, where the range of distances becomes a vibrant illustration of each individual's unique exploration within the diverse topography of their wellness journey.

```python
%%sql
SELECT Id, Day,
       MAX(TotalDistance) AS MaxDistance,
       MIN(TotalDistance) AS MinDistance,
       (MAX(TotalDistance) - MIN(TotalDistance)) AS DistanceRange
FROM smart_watch_data
GROUP BY Id, Day
ORDER BY Id, Day;
```

# Peaks of Inactivity Unveiled.
In our data panorama, this SQL panorama unfolds as an observer, gazing upon the peaks of sedentary time within 'smart_watch_data.' By aggregating the percentage of sedentary minutes against total minutes for each 'Id,' the code reveals the individuals with the highest proportion of inactive intervals. It's not just a percentage; it's a glimpse into the landscapes where stillness dominates. This code becomes a storyteller, narrating the chapters of those whose wellness journey encounters the heights of sedentary time, prompting reflection on the balance between rest and activity in the diverse terrain of daily life.

```python
%%sql
SELECT Id,
       SUM(SedentaryMinutes) AS TotalSedentaryMinutes,
       SUM(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes + SedentaryMinutes) AS TotalMinutes,
       ROUND((SUM(SedentaryMinutes) / SUM(VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes + SedentaryMinutes) * 100), 2) AS PercentageSedentary
FROM smart_watch_data
GROUP BY Id
ORDER BY PercentageSedentary DESC;
```
